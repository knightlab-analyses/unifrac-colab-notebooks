```python
%%bash

# create our conda environment and install:
# unifrac-binaries -> the c++ compiled code and shared library
# unifrac -> a python wrapper which links to the shared library
# biom-format -> the library we'll use to subsample the full 330k feature table
# time -> GNU time as colabs seems to lack it
conda create -q --name unifrac -y -c conda-forge -c bioconda unifrac-binaries=1.0.0 unifrac=1.0.0 biom-format time
```

```python
%%bash

# show the help information for the command line application

source activate unifrac
ssu
```

```python
%%bash

# pull a small test dataset, and run it. The output should indicate GPU is used
# on the first run, and cpu on the second run, then we'll verify the output is
# identical 

source activate unifrac

wget --quiet -L https://github.com/biocore/unifrac/raw/master/unifrac/tests/data/crawford.biom
wget --quiet -L https://github.com/biocore/unifrac/raw/master/unifrac/tests/data/crawford.tre

export UNIFRAC_GPU_INFO=Y
ssu -i crawford.biom -t crawford.tre -m unweighted_fp32 -o test.dm.gpu

# force CPU use
export UNIFRAC_USE_GPU=N
ssu -i crawford.biom -t crawford.tre -m unweighted_fp32 -o test.dm.cpu

# verify the output is the same
md5sum test.dm.gpu
md5sum test.dm.cpu
```

```python
%%bash

# obtain the unrarefied full 330k dataset used in the manuscript

wget --quiet -L "https://zenodo.org/record/6125920/files/330k-features.newick?download=1" -O 330k-features.newick
wget --quiet -L "https://zenodo.org/record/6126308/files/330k.tree-overlap.biom?download=1" -O 330k.tree-overlap.biom
wget --quiet -L "https://zenodo.org/record/6126147/files/metadata.tsv?download=1" -O metadata.tsv

ls -lh
```

```python
%%bash

# we do not have sufficient resources on colab to process the dataset as a whole,
# as the distance matrix requires about 700GB of memory, so let's reduce to the 
# American Gut (qiita study 10317) which represents roughly 25k samples here, 
# rarefy the data to 500 sequences per sample, and report the number of features 
# and number of samples remaining

# the table is large so rarefaction unfortunately takes a few minutes. this can be
# made faster by applying with_replacement=True in the subsample call, however there
# is still a reasonable amount of IO overhead.

source activate unifrac

python -c """
import biom, h5py
t = biom.load_table('330k.tree-overlap.biom')
f = h5py.File('330k_even500.biom', 'w')
ag = {i for i in t.ids() if i.startswith('10317')}
t = t.filter(ag).remove_empty().subsample(500)
t.to_hdf5(f, 'example')
f.close()
print(t.shape)
"""
```

```python
%%bash

# run unweighted unifrac, compute principal coordinates on the first 10 axes, 
# and measure the resources used. note that the requested output format, 
# hdf5_fp32, is one of multiple options and allows for storing both the 
# distance matrix and coordinates within the same file.
source activate unifrac

export UNIFRAC_GPU_INFO=Y
export UNIFRAC_USE_GPU=Y
$(which time) -v ssu -i 330k_even500.biom -t 330k-features.newick -m unweighted_fp32 -o result.dm -r hdf5_fp32 --pcoa 10
ls -lh result.dm
```

```python
%%bash
 
# run unweighted unifrac on the feature table using Python
source activate unifrac
 
export UNIFRAC_GPU_INFO=Y
export UNIFRAC_USE_GPU=Y

# note that we are writing directly to file, however if we used
# unifrac.unweighted_fp32, it would return an in-memory skbio.DistanceMatrix
# object
$(which time) -v python -c """
import unifrac
unifrac.unweighted_fp32_to_file('330k_even500.biom','330k-features.newick', 'result2.dm')
"""

ls -lh result2.dm

# we can also compute principal coordinates directly from the distance matrix
$(which time) -v python -c """
import unifrac
p1=unifrac.h5pcoa('result2.dm')
print(p1)
"""
```
